```python
#This Jupyter NoteBook is to analysis the traffic load matrix obtain through the execution of the programs.

import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
traffic_data = []
with open('traffic_load.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        # Filter out any empty strings and convert valid numbers to float
        traffic_data.append([float(item) for item in row if item])

# Convert to numpy array
traffic_data = np.array(traffic_data)
```

```python
# Plot traffic load heatmap
plt.imshow(traffic_data, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Traffic Load Heatmap')
plt.xlabel('Destination Node')
plt.ylabel('Source Node')
plt.show()
```

```python
# Plot time complexity (example)
times = [0.2, 0.5, 1.0, 1.5]  # execution time over different runs
nodes = [100, 200, 300, 400]  # Corresponding node sizes

plt.plot(nodes, times, marker='o')
plt.title('Time Complexity of Floyd-Warshall Algorithm')
plt.xlabel('Number of Nodes')
plt.ylabel('Execution Time (seconds)')
plt.show()
```

```python
# Total traffic per source node
total_traffic_per_node = np.sum(traffic_data, axis=1)

plt.figure(figsize=(12, 6))
plt.plot(total_traffic_per_node, marker='o', color='b', linestyle='-', label="Total Traffic per Node")
plt.title('Total Traffic Routed from Each Source Node')
plt.xlabel('Source Node')
plt.ylabel('Total Traffic Routed')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(traffic_data.shape[0])
Y = np.arange(traffic_data.shape[1])
X, Y = np.meshgrid(X, Y)
Z = traffic_data

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm')
plt.title('3D Surface Plot of Traffic Load')
plt.xlabel('Source Node')
plt.ylabel('Destination Node')
fig.colorbar(surf)
plt.show()
```
